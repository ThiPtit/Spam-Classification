```python
import os
import pandas as pd
import matplotlib as mpl
import numpy as np
from sklearn.model_selection import train_test_split
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
data = []
folder_spam = r'H:\CodePy\BaiThi2\BaiThi2\TrainData\spam'
folder_notspam = r'H:\CodePy\BaiThi2\BaiThi2\TrainData\notspam'

for filename in os.listdir(folder_spam):
    if filename.endswith('.txt'): # Chỉ đọc file có đuôi là .txt
        file_path = os.path.join(folder_spam, filename)
        with open(file_path, 'r') as file:
            content = file.read()
            data.append({'label' : 1, 'content' : content})

for filename in os.listdir(folder_notspam):
    if filename.endswith('.txt'): # Chỉ đọc file có đuôi là .txt
        file_path = os.path.join(folder_notspam, filename)
        with open(file_path, 'r') as file:
            content = file.read()
            data.append({'label' : 0, 'content' : content})
           

df = pd.DataFrame(data) 
df
```

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\W+', ' ', text)
    return text

df['tokens'] = df['content'].apply(preprocess_text)
display(df)
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if not word in stop_words]

def tokenize_text(text):
    tokens = word_tokenize(text)
    tokens = remove_stopwords(tokens)
    return tokens

df['tokens'] = df['tokens'].apply(tokenize_text)
display(df)
df['tokens'].iloc[0]
```

```python
stemmer = PorterStemmer()
def stemming(tokens):
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    return stemmed_tokens
df['tokens'] = df['tokens'].apply(stemming)
display(df)
df['tokens'].iloc[0]
```

```python
def to_string(tokens):
    s = ""
    for i in tokens:
        i += " "
        s += i
    return s
df['tokens'] = df['tokens'].apply(to_string)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Chuyển đổi dữ liệu sang dạng vector
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['tokens'])

# Hiển thị số chiều của ma trận vector
print(X.shape)
```

```python
y = df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

# Tiêu chuẩn hóa dữ liệu

scaler = StandardScaler(with_mean=False)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf = MultinomialNB()

# Huấn luyện mô hình
clf.fit(X_train, y_train)

# Dự đoán nhãn của các email trong tập kiểm tra
y_pred = clf.predict(X_test)

# Đánh giá độ chính xác của mô hình
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
